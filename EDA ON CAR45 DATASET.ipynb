```python

# Count the occurrences of each car brand
car_brand_counts = df['Brand'].value_counts()

# Select the top 10 car brands
top_10_car_brands = car_brand_counts.head(10)

# Create a bar chart
plt.figure(figsize=(10, 6))
top_10_car_brands.plot(kind='bar')
plt.title('Top 10 Most Commonly Listed Car Brands')
plt.xlabel('Car Brands')
plt.ylabel('Number of Listings')
plt.xticks(rotation=45)
plt.tight_layout()

# Annotate each bar with its value
for index, value in enumerate(top_10_car_brands):
    plt.text(index, value, str(value), ha='center', va='bottom')

plt.show()
```

```python

# Count the occurrences of each car model
model_counts = df['model'].value_counts()

# Get the top 10 most common car models
top_10_models = model_counts.head(10)

# Create a bar plot to visualize the results
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
top_10_models.plot(kind='bar', color='skyblue')
plt.title('Top 10 Most Common Car Models')
plt.xlabel('Car Model')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.tight_layout()

# Annotate the bars with data labels
for index, value in enumerate(top_10_models):
    plt.text(index, value + 5, str(value), ha='center', va='bottom')

plt.show()
```

```python
# Extract year from "year_of_man" column
df['year_of_man'] = df['year_of_man'].astype(str).str[:4]

# Count car listings for each year
year_counts = df['year_of_man'].value_counts().sort_index()

# Find the year with the highest count
max_year = year_counts.idxmax()

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(year_counts.index, year_counts.values, marker='o', color='blue')
plt.title('Number of Car Listings by Year')
plt.xlabel('Year')
plt.ylabel('Number of Listings')
plt.xticks(rotation=45)
plt.annotate(f'Highest: {max_year}', xy=(max_year, year_counts[max_year]), xytext=(-15, 30),
             textcoords='offset points', arrowprops=dict(facecolor='black', arrowstyle='wedge,tail_width=0.7'))
plt.tight_layout()
plt.grid(True)
plt.show()




```

```python
# Calculate average mileage for Foreign Used and Nigerian Used cars
avg_mileage_foreign_used = df[df['condition'] == 'Foreign Used']['mileage'].mean()
avg_mileage_nigerian_used = df[df['condition'] == 'Nigerian Used']['mileage'].mean()

# Create a bar chart
plt.figure(figsize=(8, 6))
plt.bar(['Foreign Used', 'Nigerian Used'], [avg_mileage_foreign_used, avg_mileage_nigerian_used], color=['skyblue', 'lightgreen'])
plt.title('Average Mileage Comparison between Foreign Used and Nigerian Used Cars')
plt.xlabel('Condition')
plt.ylabel('Average Mileage')
plt.tight_layout()

# Annotate each bar with its value
plt.text(0, avg_mileage_foreign_used + 5000, f'{avg_mileage_foreign_used:.0f}', ha='center', va='bottom', fontsize=12, color='black')
plt.text(1, avg_mileage_nigerian_used + 5000, f'{avg_mileage_nigerian_used:.0f}', ha='center', va='bottom', fontsize=12, color='black')

plt.show()





```

```python
# Count the distribution of transmission types
transmission_counts = df['transmission'].value_counts()

# Plotting
plt.figure(figsize=(8, 6))
sns.barplot(x=transmission_counts.index, y=transmission_counts.values, palette="viridis")
plt.title('Distribution of Transmission Types')
plt.xlabel('Transmission Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
# Annotate the bars with data labels
for index, value in enumerate(transmission_counts):
    plt.text(index, value, str(value), ha='center', va='bottom', fontsize=10, color='black')


plt.show()
```

```python
# Count the number of car listings per region
region_counts = df['region'].value_counts()

# Plotting
plt.figure(figsize=(10, 8))
sns.barplot(x=region_counts.values, y=region_counts.index, palette="mako")
plt.title('Number of Car Listings by Region')
plt.xlabel('Number of Listings')
plt.ylabel('Region')
plt.tight_layout()
# Annotate the bars with data labels
for index, value in enumerate(region_counts):
    plt.text(value, index, str(value), ha='left', va='center', fontsize=10, color='black')


plt.show()
```

```python
# Count the distribution of condition types
condition_counts = df['condition'].value_counts()

# Plotting
plt.figure(figsize=(8, 6))
sns.barplot(x=condition_counts.index, y=condition_counts.values, palette="viridis")
plt.title('Distribution of condition Types')
plt.xlabel('condition Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
# Annotate the bars with data labels
for index, value in enumerate(condition_counts):
    plt.text(index, value, str(value), ha='center', va='bottom', fontsize=10, color='black')

plt.show()
```

```python
# Count the distribution of fuel types
fuel_counts = df['fuel_type'].value_counts()

# Plotting
plt.figure(figsize=(8, 6))
sns.barplot(x=fuel_counts.index, y=fuel_counts.values, palette="viridis")
plt.title('Distribution of fuel Types')
plt.xlabel('fuel Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
# Annotate the bars with data labels
for index, value in enumerate(fuel_counts):
    plt.text(index, value, str(value), ha='center', va='bottom', fontsize=10, color='black')

plt.show()
```

```python
# Count the distribution of selling conditions
selling_condition_counts = df['selling_cond'].value_counts()

# Count the distribution of bought conditions
bought_condition_counts = df['bought_cond'].value_counts()

# Plotting
plt.figure(figsize=(12, 5))

# Plot for Selling Conditions
plt.subplot(1, 2, 1)
plt.bar(selling_condition_counts.index, selling_condition_counts.values, color='skyblue')
plt.title('Distribution of Selling Conditions')
plt.xlabel('Selling Condition')
plt.ylabel('Count')
plt.xticks(rotation=45)
for index, value in enumerate(selling_condition_counts):
    plt.text(index, value, str(value), ha='center', va='bottom', fontsize=10, color='black')

# Plot for Bought Conditions
plt.subplot(1, 2, 2)
plt.bar(bought_condition_counts.index, bought_condition_counts.values, color='lightgreen')
plt.title('Distribution of Bought Conditions')
plt.xlabel('Bought Condition')
plt.ylabel('Count')
plt.xticks(rotation=45)
for index, value in enumerate(bought_condition_counts):
    plt.text(index, value, str(value), ha='center', va='bottom', fontsize=10, color='black')

plt.tight_layout()
plt.show()





```
